```python
import datetime as datetime
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
gamestop = pd.read_csv('GME.csv', index_col='Date',parse_dates=True)
```

```python
gamestop.head(10)
```

```python
gamestop.index.max()
```

```python
# daily to yearly
gamestop.resample(rule='A').mean()
```

```python
gamestop['Close'].resample('M').mean().plot()
plt.show()
```

```python
gamestop['Volume'].resample('M').plot()
plt.show()
```

```python
gamestop['Close'].resample('D').mean().plot()
plt.show()
```
