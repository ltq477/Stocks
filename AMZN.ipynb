```python
import datetime as datetime
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
amzn = pd.read_csv('amzn.csv', index_col='Date',parse_dates=True)
```

```python
amzn.head(10)
```

```python
amzn.index.max()
```

```python
# daily to yearly
amzn.resample(rule='A').mean()
```

```python
amzn['Close'].resample('M').mean().plot()
plt.show()
```

```python
amzn['Volume'].resample('M').plot()
plt.show()
```

```python
amzn['Close'].resample('D').mean().plot()
plt.show()
```
